```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_excel('Data_Train.xlsx')
test=pd.read_excel('Test_set.xlsx')
```

```python
train.head()
```

```python
test.head()
```

```python
final_df=pd.concat([train,test])
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0].astype(int)
final_df['month']=final_df['Date_of_Journey'].str.split('/').str[1].astype(int)
final_df['year']=final_df['Date_of_Journey'].str.split('/').str[2].astype(int)
```

```python
final_df.head()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df.isnull().sum()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'NaN	':1})
```

```python
final_df.head(1)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.isnull().sum()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].fillna(1)
```

```python
final_df.isnull().sum()
```

```python
final_df.info()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.head()
```

```python
# Extract hours (if present), fill NaN with 0, convert to int
hours = final_df['Duration'].str.extract(r'(\d+)h')[0].fillna(0).astype(int)

# Extract minutes (if present), fill NaN with 0, convert to int
minutes = final_df['Duration'].str.extract(r'(\d+)m')[0].fillna(0).astype(int)

# Total duration in minutes
final_df['Duration_in_Minutes'] = hours * 60 + minutes
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
final_df['Airline']=lb.fit_transform(final_df['Airline'])
final_df['Source']=lb.fit_transform(final_df['Source'])
final_df['Destination']=lb.fit_transform(final_df['Destination'])
final_df['Additional_Info']=lb.fit_transform(final_df['Additional_Info'])
```
