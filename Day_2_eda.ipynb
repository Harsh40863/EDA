```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test=pd.read_csv('test.csv')
test.head()
```

```python
train=pd.read_csv('train.csv')
train.head()
```

```python
df = pd.concat([train, test])
df.head()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Gender']=pd.get_dummies(df['Gender'],drop_first=True).astype(int)
```

```python
df.head()
```

```python
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'55+':2,'26-35':3,'46-50':4,'51-55':5,'36-45':6,'18-25':7})
```

```python
df.head()
```

```python
df['City_Category'].unique()
```

```python
city=pd.get_dummies(df['City_Category'],drop_first=True).astype(int)
```

```python
city.head()
```

```python
df=pd.concat([df,city],axis=1)
```

```python
df.head()
```

```python
df.drop(['City_Category'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df.isnull().sum()
```
